```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/Yunan/Desktop/IS445/store_data.csv',encoding='latin1')
```

```python
df.columns
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

#Add a new column to show only the year
df['Year'] = df['Order Date'].apply(lambda x: x.year)
```

```python
sales = df.groupby('Year')['Sales'].sum()
```

```python
sales_by_year = df.groupby('Year')['Sales'].sum()
ax = sales_by_year.plot(kind='line', figsize=(10, 5),color='green')
ax.set_title('Total Sales by Year')
ax.set_xlabel('Year')
ax.set_ylabel('Sales')
plt.show()
```

```python
sales_by_region = df.groupby('Region')['Sales'].sum()

# Create a pie chart to show the distribution of sales by region
ax = sales_by_region.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 5))
ax.set_title('Sales by Region')
plt.show()
```

```python
sales_by_category_region = df.groupby(['Category', 'Region'])['Sales'].sum()

# Pivot the data to create a bar chart
sales_by_category_region = sales_by_category_region.unstack()

# Create the bar chart
ax = sales_by_category_region.plot(kind='bar', stacked=True, figsize=(10, 6))
ax.set_title('Sales by Category and Region')
ax.set_xlabel('Category')
ax.set_ylabel('Sales')
plt.show()
```
